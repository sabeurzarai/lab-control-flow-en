## Words

```python
from IPython.lib.display import IFrame

words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
for i, word in enumerate(words):
    print(word.upper())

```

**Create a new list containing only words with 5 or more letters**

```python
words_new = []
for i, word in enumerate(words):
    if len(word) >= 5:
        words_new.append(word)

print(words_new)
```

**Print the first word starting with "t"**

```python
for i, word in enumerate(words):
    if words[i].startswith("t"):
        print(words[i])
        break
```

## Numbers

**Create a list containing the square of every number from 1 to 10**

```python
squares = []
for i in range(1, 11):
    squares.append(i ** 2)
print(squares)
```

**Print a list containing the square of every odd number from 1 to 10**

```python
odd_sequence = []
for i in range(1,10):
    if i % 2 != 0:
        odd_sequence.append(i ** 2)
        i += 1
print(odd_sequence)
```

**Create a list with the squares of all multiples of 8 below 1000**

```python
mulptiple_of_eight_squence = []
for i in range(1,1000):
    if i % 8 == 0:
        mulptiple_of_eight_squence.append(i ** 2)
print(mulptiple_of_eight_squence)
```

## People

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]
```

**How many people are there?**

```python
print(len(people))
```

**How many people have kids**?

```python
for i, person in enumerate(people):
    if people[i]["n_kids"] > 0:
        i += 1
print(i)
```

**How many kids do they have in total?**

```python
number_of_kids = 0
for i, person in enumerate(people):
    if people[i]["n_kids"] > 0:
        number_of_kids =+ people[i]["n_kids"]
print(number_of_kids)
```

**In a year's time, names ending with "a" will have an extra kid. Create a list of dictionaries with people's info in a year's time**

```python
in_yeartime_list_of_people = []
for i, person in enumerate(people):
    people[i]["age"] = people[i]["age"] + 1
    if people[i]["name"].endswith("a"):
        people[i]["n_kids"] = people[i]["n_kids"] + 1
    in_yeartime_list_of_people.append(people[i])

print(in_yeartime_list_of_people)
```
